```python
import numpy as np 
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc
```

```python
y_true = np.array([1,1,0,0,1])   #Actual labels
y_scores=np.array([0.95,0.85,0.70,0.40,0.30])    #probabilty
```

```python
fpr, tpr, tnresholds =roc_curve(y_true, y_scores)
roc_auc = auc(fpr, tpr)

print("auc:", roc_auc)
```

```python
plt.figure()
plt.plot(fpr, tpr, label=f'ROC curve (AUC = {roc_auc:.2f})')
plt.plot([0,1],[0,1], linestyle='--')
plt.xlabel("False Positive Rate") 
plt.ylabel("True Positive Rate")
plt.title('ROC curve')
plt.legend()
plt.show()
```
